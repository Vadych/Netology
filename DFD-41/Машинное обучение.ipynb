# Предсказываем стоимость мед страховки



## Загужаем необходимые библиотеки

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np
```

## Загружаем данные

Собраны данные:
- age: возраст
- sex: пол
- bmi: индекс массы тела
- children: количество детей, охваченных медицинским страхованием / количество иждивенцев
- smoker: курение
- region: регион (northeast, southeast, southwest, northwest).
- charges: индивидуальные медицинские расходы (его и хотим предсказать)

```python
# Загрузите данные из файла insurance.csv в переменную df
df = pd.read_csv('insurance.csv')
```

```python
df.head()
```

## Смотрим статистику, что нет пропусков и отсуствующих значений

```python
df.info()
```

```python
df.describe()
```

## Преобразуем строковые данные

```python
# Заменяем пол и курение на числа
df['sex']=df['sex'].map({'male':1, 'female':0})
df['smoker']=df['smoker'].map({'yes':1,'no':0})
```

```python
df.head()
```

```python
# Заменяем регион на набор отдельных колонок (is_southwest, is_southeast и тп)
df = pd.get_dummies(df, columns=['region'])
df.head()
```

## Формируем признаки и целевую переменную

```python
## Сформируйте признаки и целевую переменную
```

```python
X, y = df.drop('charges', axis=1), df['charges']
```

## Разделяем данные на выборку для обучения/проверки

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## Создаем и обучаем модель линейной регресии

```python
# Создайте и обучите модель
lr = LinearRegression()
lr.fit(X, y)
lr.coef_
```

## Получаем предсказание и оцениваем качество

```python
# Получите предсказание
lr.predict(X_test)
```

```python
# Оцените качество, при помощи метода mean_squared_error для тестовой выборки
```

```python
np.sqrt(mean_squared_error(y_test, lr.predict(X_test)))
```

# Задание со звездочкой

```python
df_cash = pd.read_csv('cash.csv', sep=',', decimal=',')
df_cash.info()
```

```python
# Заполняем пропуски

df_cash['m-2'].fillna(df_cash['m-2'].median(), inplace=True)
df_cash['m-1'].fillna(df_cash['m-1'].median(), inplace=True)
df_cash['next'].fillna(df_cash['next'].mean(), inplace=True)
df_cash.info()
```

```python
# Делим выборку
X = df_cash[['m-2', 'm-1', 'm']]
y = df_cash['next']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Обучаем модель
lr_cash = LinearRegression()
lr_cash.fit(X_train, y_train)
lr_cash.coef_
```

```python
# Оценка качества
pd.DataFrame([y_test.values, lr_cash.predict(X_test)]).T
```

```python
np.sqrt(mean_squared_error(y_test, lr_cash.predict(X_test)))
```
