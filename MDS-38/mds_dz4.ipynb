Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
x
```

```python
f = sym.sin(2*x+1)**5
d1 = sym.diff(f)
d2 = sym.diff(d1)
arg = np.arange(-5, 5.01, 0.01)
y = [f.evalf(subs={x: t}) for t in arg]
y1 = [d1.evalf(subs={x: t}) for t in arg]
y2 = [d2.evalf(subs={x: t}) for t in arg]

plt.plot(arg, y, label = 'Функция', color = 'black')
plt.plot(arg, y1, label = 'Первая', color = 'green')
plt.plot(arg, y2, label = 'Вторая', color = 'red')
plt.legend()
plt.xlim(-5, 5)
plt.show()
```
