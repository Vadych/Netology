### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```python
p = np.full((3, 2), 0.5) # Матрица вероятности выпасть орлу или решки для каждой монеты
p_A1 = p[0] @ p[1] # Вектора - вероятности выпадения орла и решки на соответсвующей монете
p_A2 = p[1] @ p[2] 
p_A3 = p[0] @ p[2]
p_All = p.prod(axis=0).sum() # Вероятность, что все монеты упали одной стороной вверх
print(f'Вероятности: A1 = {p_A1}, A2 = {p_A2}, A3 = {p_A3}, все = {p_All}')
print(f'Вероятности: A1 и А2 = {p_A1*p_A2}, A2 и А3 = {p_A2*p_A3}, A1 и A3 = {p_A1*p_A3}, A1, А2 и А3 = {p_A1*p_A2*p_A3}')
```

Не трудно заметить, что попарно события А1, А2 и А3, а так же все 3 события всместе дают исход - все монеты лежат вверх одной стороной. Вероятность такого исхода равна 0.25

Попарное произведение вероятностей событий А тоже дает 0.25, что говорит об их независимости. 

Произведение вероятности всех событий А дает 0.125, что говорит о зависимости событий, т.к. это число не равно 0.25

Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения. 

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```python
from scipy.stats import norm, expon, t
import matplotlib.pyplot as plt
x = np.linspace(-4, 4, 200)
```

```python
def show_distribution(stat, parameter, p_list, x = x):
    for par in p_list:
        plt.plot(x, stat.pdf(x, **{parameter: par}), label=f'{parameter} = {par:.2f}')
    plt.legend()
    plt.show()
```

```python
show_distribution(norm, 'loc', [-2, -1, 0, 1, 2])
```

```python
show_distribution(norm, 'scale', [0.5, 1, 1.5, 2, 2.5])
```

```python
show_distribution(expon, 'loc', [0.5, 1, 1.5, 2, 2.5])
```

```python
show_distribution(expon, 'scale', [0.5, 1, 1.5, 2, 2.5])
```

```python
show_distribution(t, 'df', [1, 5, 10, 20, 100])
```
