## Задание 1
Переведите содержимое файла purchase_log.txt* в словарь purchases вида:
{‘1840e0b9d4’: ‘Продукты’, …}

```python
import json

purchases = {}
with open('purchase_log.txt') as f:
  f.readline()
  purchases = {row['user_id']:row['category'] for row in map(json.loads, f.readlines())}
```

## Задание 2

Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки, если покупка была, сам файл visit_log.csv* изменять не надо. Запишите в файл funnel.csv визиты из файла visit_log.csv*, в которых были покупки с указанием категории.

```python
with open('visit_log.csv') as f_in:
  with open('funnel.csv', 'w', encoding = 'utf-8') as f_out:
    f_out.write(f_in.readline()[:-1] + ',category\n')
    for line in f_in:
      record = line.strip().split(',')
      if record[0] in purchases:
        f_out.write(','.join(record) + f',{purchases[record[0]]}\n')
```
