# Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df.info()
```

```python
df[df['surgery'].isna()]
```

```python
# Один пропуск. Заполним модой
df.loc[df.surgery.isna(), 'surgery'] = df['surgery'].mode()[0]
```

```python
df[df['rectal_temperature'].isna()]
```

```python
# Видно, что есть строки, где пропущено 3 и более столбцов. Их логичней удалить
df = df.dropna(thresh=4)
df.reset_index(drop=True, inplace=True)
df
```

```python
# Оставшиеся линейные признаки заполним средним значением, 
# уже имеющейся в нашем наборе.
# Изначально хотел заполнить с помощью np.random.normal, но не справился.
# Надеюсь на практике будет показано как это сделать

for col in ['rectal_temperature', 'pulse', 'respiratory_rate']:
    df[col].fillna(df[col].mean(), inplace=True)

df
```

```python
# Было предположение, что температура конечностей должна быть связана с ректальной. 
# Не подтвердилось 

df.pivot_table(
    index='temperature_of_extremities',
    aggfunc='mean',
    values='rectal_temperature'
)
```

```python
# Оставшиеся категориальные признаки заполним модой
for col in ['temperature_of_extremities', 'pain', 'outcome']:
    df[col].fillna(df[col].mode()[0], inplace=True)

df
```
