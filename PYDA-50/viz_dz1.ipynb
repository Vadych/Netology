### Год

```python
# Сохраняем правильные года
years = ['1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002',
         '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010',
         '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018',
         '2019', '2020', '2021', '2022', '2023']
```

Пытаемся найти место с правильным годом

```python
def year_to_int(row):
    """
    В результате исследований выяснилось, что год может находиться в некоторых других столбцах.
    Эта функция возвращает правильный год для строки
    """
    if row.year in years: # Год в правильном месте
        return int(row.year)
    for val in row.values: # Пытаемся найти год в другом поле
        if val in years:
            return int(val)
    for val in row.values: # Пытаемся найти год как строку типа 06/2023    
        res = re.findall(r'\d\d/(\d\d\d\d)', val)
        if len(res) != 0:
            return int(res[0])
    
```

```python
car['int_year'] = car.apply(year_to_int, axis=1)
```

```python
car.int_year.unique()
```

### Тип топлива 

```python
fuel_type = ['Petrol', 'Diesel', 'Hybrid', 'Electric', 'LPG', 'CNG', 'Diesel Hybrid',
             'Other', 'Hydrogen', 'Ethanol']
```

```python
car.fuel_type.value_counts().head(20)
```

```python
def fuel(row):
    if row.fuel_type in fuel_type:
        return row.fuel_type
    for val in row.values:
        if val in fuel_type:
            return val
    
```

```python
car['fuel_type'] = car.apply(fuel, axis=1)
```

Для электрокаров похоже указан или пробег на полной заправке или затрачиваемая мощность

```python
def fuel_consumption(row):
    """
    Возвращает расход, если не удалось преобразовать, то Null
    """
    fc = str(row.fuel_consumption_l_100km)
    fc = fc.replace(',', '.')
    val = re.findall(r'(\d+\.?\d?) l/100 km', fc)
    try:
        return float(val[0])
    except:
        pass
```

```python
car['fuel_consumption'] = car.apply(fuel_consumption, axis=1)
```

```python
car[car.fuel_consumption.isna()]
```

Очень много пропусков. Решил, что восстанавливать не имеет смысла. Будем жить с этим
